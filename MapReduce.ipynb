```python
import glob
from bs4 import BeautifulSoup
from pyspark.sql import SparkSession
```

```python
from joblib import Parallel, delayed
```

```python
shared_set = set()
data = []
def FindLink(file_name):
  with open(file_name,'r',encoding='utf-8',errors='ignore') as fp:
    print(file_name)
    soup = BeautifulSoup(fp, "html.parser")
    soup.find_all("a", href=True)
    for link in soup.find_all("a", href=True):
      data.append(link['href'])

Parallel(n_jobs=8, require='sharedmem')(delayed(FindLink)(file_name) for file_name in glob.glob("/content/*.html"))
```

```python
data=[]
for file_name in glob.glob("/content/*.html"):
  with open(file_name,'r',encoding='utf-8',errors='ignore') as fp:
    soup = BeautifulSoup(fp, "html.parser")
    soup.find_all("a", href=True)
    for link in soup.find_all("a"):
      data.append(link.get('href'))
      if (link.get('href') == '#'):
        print(file_name)
        
```

```python
print(len(data))
```

```python
spark = SparkSession.builder.appName('test').getOrCreate()

rdd=spark.sparkContext.parallelize(data)

rdd2=rdd.map(lambda x: (x,1)) \
        .reduceByKey(lambda a, b: a + b) \
        .sortBy(lambda a:[a[1],len(a[0])] ,False)

for element in rdd2.filter(lambda x:x[1] >5).collect():
    print(element)
```
